```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
BostonTrain = pd.read_csv("boston_test.csv")
```

```python
BostonTrain.head()
```

```python
BostonTrain.info()
BostonTrain.describe()
```

```python
BostonTrain.isnull()
# Count total missing values per column
print(BostonTrain.isnull().sum())
```

```python
BostonTrain = BostonTrain.dropna()
```

```python
print(BostonTrain.isnull().sum())
```

```python
plt.subplots(figsize=(12,8))
sns.heatmap(BostonTrain.corr(), cmap = 'RdGy')
# Display the plot
plt.show()
```

```python
sns.pairplot(BostonTrain, vars = ['RM', 'ZN', 'BLACK', 'DIS', 'CHAS'])
# Show the plot
plt.show()
```

```python
X = BostonTrain[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'BLACK', 'LSTAT']]
y = BostonTrain['RAD']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)
```

```python
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
sns.histplot((y_test - predictions), bins=50, kde=True)

plt.xlabel('Residuals')
plt.title('Distribution of Residuals')
plt.show()
```

```python
coefficients = pd.DataFrame(lm.coef_,X.columns)
coefficients.columns = ['coefficients']
coefficients
```
